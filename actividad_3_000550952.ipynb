```python
import math
frecuencia_señal = int(input("Ingrese la frecuencia de la señal: "))
MEDIA = 1000
DESVIACION = 300
PERCENT = 0.3
pr_acumulada = 0
for f in range(frecuencia_señal + 1):
    resultado = (1 / (DESVIACION * math.sqrt(2 * math.pi))) * math.exp(-(1/2) * ((f - MEDIA) / DESVIACION) ** 2)
    pr_acumulada += resultado
if pr_acumulada > PERCENT:
    print(pr_acumulada)
    print("La señal pasa el filtro.")
else:
    print(pr_acumulada)
    print("La señal es bloqueada por el filtro.")
```
